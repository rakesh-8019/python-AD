# simple linear regression

```python
ml_v=smf.ols('MPG~VOL',data=cars).fit()
np.round(ml_v.rsquared,2)
```

```python
ml_v.params
```

```python
ml_w=smf.ols('MPG~WT',data = cars).fit()
ml_w.rsquared
```

```python
ml_w=smf.ols('MPG~MPG',data = cars).fit()
ml_w.rsquared
```

```python
ml_w=smf.ols('MPG~HP',data = cars).fit()
ml_w.rsquared
```

```python
ml_w=smf.ols('MPG~SP',data = cars).fit()
np.round(ml_w.rsquared,2)
```

```python
ml_w.params
```

```python
model = smf.ols('MPG~WT+VOL+SP+HP',data=cars).fit()
model.rsquared
```

```python
model.params
```
